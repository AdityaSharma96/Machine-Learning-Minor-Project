```python
import pandas as pd
import numpy as np


import matplotlib.pyplot as plt
from pylab import rcParams


from scipy import stats
from sklearn.metrics import mean_squared_error as MSE


from sklearn.model_selection import train_test_split
from sklearn import preprocessing
from sklearn.linear_model import Lasso
```

```python
trainingdata=pd.read_csv(r"C:\Users\HP\Desktop\project\train.csv")
testdata=pd.read_csv(r"C:\Users\HP\Desktop\project\test.csv")
```

```python
%matplotlib inline
rcParams['figure.figsize']=10,4
```

```python
print("Training data shape (Rows,Coloumns):",trainingdata.shape)
print("Test data shape (Rows,Coloumns):",testdata.shape)
```

```python
pd.options.display.max_columns=None
display(trainingdata.head())
display(testdata.head())
```

```python
trainingdatacolumns=list(trainingdata)

for c in trainingdatacolumns:
    if trainingdata[c].dtype=='O':
        trainingdata[c].fillna(value='Missing',inplace=True)
    else:
        trainingdata[c].fillna(0,inplace=True)
        
        
        
#same for test data

testdatacolumns=list(testdata)

for f in testdatacolumns:
    if testdata[f].dtype=='O':
        testdata[f].fillna(value='Missing',inplace=True)
    else:
        testdata[f].fillna(0,inplace=True)

pd.options.display.max_columns=None
display(trainingdata.head())
display(testdata.head())
```

```python
#transforming discrete values to columns with 1 and 0s
trainingdata=pd.get_dummies(trainingdata)

#doing same for competitive data
testdata=pd.get_dummies(testdata)


pd.options.display.max_columns=None
display(trainingdata.head())
display(testdata.head())
```

```python
print("Training data shape (Rows,Coloumns):",trainingdata.shape)
print("Test data shape (Rows,Coloumns):",testdata.shape)
```

```python
#dropping features that are not present in both sets and making same value of shape of test data set and trianing data set

sp=trainingdata['SalePrice']

missingfeatures=list(set(trainingdata.columns.values)-set(testdata.columns.values))
trainingdata=trainingdata.drop(missingfeatures,axis=1)

missingfeatures=list(set(testdata.columns.values)-set(trainingdata.columns.values))
testdata=testdata.drop(missingfeatures,axis=1)

print("Training data shape (Rows,Coloumns):",trainingdata.shape)
print("Test data shape (Rows,Coloumns):",testdata.shape)
```

```python
x_train,x_test,y_train,y_test=train_test_split(trainingdata,sp,random_state=0)
```

```python
model = Lasso(alpha=298.4).fit(x_train,y_train)
print("Train score: ",model.score(x_train,y_train),"\nTest score: ",model.score(x_test,y_test))
print("Number of features used: {}".format(np.sum(model.coef_ !=0)))
```

```python
result=pd.DataFrame(model.predict(testdata),columns=['SalesPrice'],index=testdata.index)
display(result.head())
```

```python
result.to_csv("result_model.csv")
```
